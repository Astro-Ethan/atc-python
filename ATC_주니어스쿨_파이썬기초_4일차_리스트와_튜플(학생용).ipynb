**튜플은 괄호를 생략해도 된다?**

튜플은 괄호를 생략할 수도 있습니다. 그렇지만 항상 (  )를 생략할 수는 없어요.

a, b, c = 1, 2, 3

a, b, c = 1, 2

위 둘 중 오류가 나는 것은 아래줄의 a, b, c = 1, 2입니다. 왜냐하면 a, b, c에 각각 대응되는 값이 없기 때문입니다.

 a, b, c = 1, 2, 3은 각각 a=(1), b=(2), c=(3)이라는 튜플이 생성된 것이구요.

 그럼 이런 성질을 활용해 아래 퀴즈의 답을 예상해보고 직접 결과를 확인해보세요.
